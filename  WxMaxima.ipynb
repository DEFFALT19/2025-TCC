```python
#Separar para que fique uma EDO linear de 2° grau nao homogenea


lhs = sp.laplace_transform(sp.diff(u, x, 2), t, s, noconds=True)
rhs = (-T1 + s * sp.laplace_transform(u, t, s, noconds=True)) / a**2

# Assume initial condition u(x,0) = T1
u0 = u.subs(t, 0)  # This is u(x,0)
rhs = rhs.subs(u0, T1)

rhs_expanded = rhs.expand()

rhs_expanded
```

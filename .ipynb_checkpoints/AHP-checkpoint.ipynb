## 1) Introdução

```python
#O objetivo é reproduzir o problema contextualizado de Analytic Hierarchy Process (AHP) de Métodos de Decisão Multi Critério (MCDM) realizado pelo artigo 
#"A ANÁLISE MULTICRITÉRIO NA TOMADA DE DECISÃO - O Método Analítico Hierárquico de T.L Saaty" de Bruno Miguel da Cruz Jordão e Susete Rodrigues Pereira na linguagem Python
```

## 2) Problema

```python
# No problema, um engenheiro civil recebeu duas ofertas de emprego (emprego 1 e 2). Ele reside atualmente na cidade de Coinbra, mas está disposta a mudar de cidade, caso as vantagens 
#oferecidas por outros fatores sejam melhores.
# Para realizar a escolha do melhor emprego, ele considerou os seguintes fatores ou critérios: salário, oportunidade de progresso profissional, localização e custo de vida no local de trabalho.

# Os critérios dos empregos são os seguintes:
# - Emprego 1
# salário: 14.000,00 €
# oportunidade de progresso profissional: Alta
# localização: Lisboa
# custo de vida no local de trabalho: Alto

# - Emprego 2
# salário: 21.000,00 €
# oportunidade de progresso profissional: Baixa
# localização: Coimbra
# custo de vida no local de trabalho: Médio
```

## 3) Construção das Matrizes de Comparação


```python
#C = criterios
#C1 = criterio 1 (slário)
# os valores mostrados são as preferências (de 1 a 9 representando menor a maior fator de importância)
#após de se ter construído a hierarquia, deve-se fazer uma comparação par a par de cada elemento no nível 
#hierárquico dado, criando-se uma matriz de decisão quadrada.

#Cm° -> Critério Número

C1 = ([1, frac(1,6)], [6,1])
#C1 = array(C1)

C2 = ([1,8],[frac(1,8),1])
#C2 = array(C2)

C3 = ([1,frac(1,5)],[5,1])
#C1 = array(C3)

C4 = ([1,4],[frac(1,4),1])
#C4 = array(C4)

print(array(C1).shape)
print(C1)
```

## 4) Normalizar as matrizes

```python
# dividir cada elemento da matriz pela soma da coluna a que pertence, de tal forma que 
# a soma de todos os seus elementos seja igual a 1

def N(C):

    n1 = C[0][0] + C[1][0]
    n2 = C[0][1] + C[1][1]

    CN = [[frac(C[0][0],n1), frac(C[0][1],n2)], 
          [frac(C[1][0],n1), frac(C[1][1],n2)]]

    return CN

print(array(N(C1)).shape)

CN1 = N(C1)
print(CN1)
print(array(CN1).shape)
print(N(C1))

#obs: colocar n1 e n2 no return para verificar se estao dando os valores certos, mas nao se esqueça de retiralos depois
           
```

```python
#determinar matriz normal de cada critério
C1N = N(C1)
C2N = N(C2)
C3N = N(C3)
C4N = N(C4)

#matrizes normalizadas
print(C1N, C2N, C3N, C4N)

#tamanhos
print(array(C1N).shape)

#se a soma de todos o elementos em cada coluna nas matrizes normalizada for igual a 1, está correto
print(C1N[0][0] + C1N[1][0])
```

## 5) Média de cada critério

```python
# converte-se as fracções em decimais e econtra-se a média aritimética de cada
# linha da matriz normalizada. O vector representado um dado critério.

def M(CN):

    m1 = float((CN[0][0] + CN[0][1])/2)
    m2 = float((CN[1][0] + CN[1][1])/2)

    CM = ([m1], [m2])
    return CM
```

```python
#teste 2

float((C1N[1][0] + C1N[1][1])/2)

print(array(M(C1N)).shape)
print(M(C1N))
```

```python
#determinar média de cada critério
# Cn°M -> Critério Número Médio
C1M = M(C1N)
C2M = M(C2N)
C3M = M(C3N)
C4M = M(C4N)

#print(C1M, C2M, C3M, C4M)
print(C1M, C2M, C3M, C4M)
print(array(C1M).shape)
```

## 6) Construir matriz de prioridade

```python
# Com os vectores encontrados na etapa anterior, deve-se
# constuir uma matriz. As linhas constituirão as alternativas e as colunas os critérios

#MP -> Matriz Prioridade
#CMP -> Construção Matriz Prioridade

def CMP(C1M, C2M, C3M, C4M):

    #linha 1    
    cmp1 = [C1M[0][0], C2M[0][0], C3M[0][0], C4M[0][0]]

    #linha 2
    cmp2 = [C1M[1][0], C2M[1][0], C3M[1][0], C4M[1][0]]

    MP = [cmp1, cmp2]
    return MP
```

```python
MP = CMP(C1M, C2M, C3M, C4M)
print(MP)
print(array(MP).shape)
```

```python
#ou seja isso é a % de preferencia de cada critério, por exemplo, no criterio 1, salário, o emprego 1 tem 14.28% de preferência enquanto o emprego 2 tem 85.71%
```

## 7) Construir a matriz de comparação dos critérios

```python
# Depois de obtido o vector de prioridades ou de impacto das alternativas 
# sob cada critério continua-se com o nível dos critérios. Construir uma matriz de comparação
# dos critérios e repetir as etapas 4 e 5 para a classificação par a par.
# O resultado será um vector que contém a média das preferências de cada critério.


# MCC -> matriz comparação de critérios
# o autor decidiu um valor que representa o nivel de relevância a cada critério
# C1 = 1, C2 = 7, C3 = 3, C4 = 2
# Ordem das linhas e colunas: C1, C2, C3, C4

MCC = [[1, frac(1,7), frac(1,3), frac(1,2)],
       [7, 1, 5, 5],
       [3, frac(1,5), 1, 3],
       [2, frac(1,5), frac(1,3), 1]]

print(MCC)
print(array(MCC).shape)




```

```python
# etapa 4: normalizar a matriz
n1 = MCC[0][0] + MCC[1][0] + MCC[2][0] + MCC[3][0]
n2 = MCC[0][1] + MCC[1][1] + MCC[2][1] + MCC[3][1]
n3 = MCC[0][2] + MCC[1][2] + MCC[2][2] + MCC[3][2]
n4 = MCC[0][3] + MCC[1][3] + MCC[2][3] + MCC[3][3]

CN = [[frac(MCC[0][0],n1), frac(MCC[0][1],n2), frac(MCC[0][2],n3), frac(MCC[0][3],n4)],
      [frac(MCC[1][0],n1), frac(MCC[1][1],n2), frac(MCC[1][2],n3), frac(MCC[1][3],n4)],
      [frac(MCC[2][0],n1), frac(MCC[2][1],n2), frac(MCC[2][2],n3), frac(MCC[2][3],n4)],
      [frac(MCC[3][0],n1), frac(MCC[3][1],n2), frac(MCC[3][2],n3), frac(MCC[3][3],n4)]]

print(CN)

#verificar soma de colunas

print(CN[0][0]+CN[1][0]+CN[2][0]+CN[3][0])
print(array(CN).shape)
```

```python
#etapa 5: media dos critérios

m1 = float((CN[0][0] + CN[0][1]) + CN[0][2] + CN[0][3])/4
m2 = float((CN[1][0] + CN[1][1]) + CN[1][2] + CN[1][3])/4
m3 = float((CN[2][0] + CN[2][1]) + CN[2][2] + CN[2][3])/4
m4 = float((CN[3][0] + CN[3][1]) + CN[3][2] + CN[3][3])/4

print(m1, m2, m3, m4)

CM = [[m1], [m2], [m3], [m4]]
print(CM)
print(array(CM).shape)
print(len(CM))

```

## 8) Obter o resultado

```python
# Multiplicar a matriz obtida na etapa 6 pelo vector da média obtido na etapa 7.
# O resultado será um vector que contém a quantificação final de cada alternativa.

MP1 = array(MP)
print(MP1)
print(MP1.shape)

CM1 = array(CM)
print(CM1)
print(CM1.shape)
```

```python
#resultado

R = dot(MP1, CM1)
print(R)
print(R.shape)

```

```python
#ou seja, no final levando em conta todos os valores de prefência dos critérios com a matriz de comparação de critérios,
# obtemos os valores preferência para cada escolha, sendo 67.92% para o emprego 1, e 32.07% para o emprego 2, mostrando que com as escolhas 
# de preferências escolhidas pelo autor (tomador de decisão) para cada critério levado em consideração, a melhor escolha para o problema contextualizado é o primeiro.
```

## 9) Calcular a coerência

```python
# O calculo da coerência deve ser tido em consideração para todos as matrizes binários do problema,
# mas só se justifica para matrizes de dimensão igual ou superiores a 3x3.
# Neste caso, só temos uma matriz de dimensão aceitável (4x4), que é a da comparação entre critérios.
# Esta etapa contem 4 passos.


#Matriz de comparção entre critérios
print(MCC)
print(array(MCC).shape)
```

### 9.1) Determinar a totalização das entradas

```python
# Resume-se o produto da matriz inicial com o respectivo vector da média, obtido na etapa 7
# que está abaixo representado.

#matriz comparação dos criterios

print(CM)

print('---------------------------------')

#totalização das entradas 

R1 = dot(MCC,CM)
print(R1, array(R1).shape)
```

### 9.2) Determinar o lambda maximo

```python
# Considera-se o n° de critérios, que nest caso são 4. Divide-se o vector do total
# de entradas pelo vector da média e obtêm-se um novo vector.
# Deste vector resultante, somam-se as suas parcelas (4) e divide-se pelo
# n° de critérios (4), e resultado é o lambda máximo.

#DL -> determinar lambda

DL = array(R1) / array(CM)
print(DL)
```

```python
#calculo do lambda max
# media dos criterios da matriz encontrada anteriormente

#LM -> Lambda max
LM = sum(DL)/len(DL)
print(LM) 
```

### 9.3) Calcular o índice de coerência, IC:

```python
# IC = (LM - n)/(n - 1), sendo n o número de criterios

n = 4
IC = (LM - n)/(n - 1)
print(IC)
```

### 9.4) Comparar entre indices:

```python
# Comparar o Índice de Coerência, IC com o Índice Aleatório, IA e verificar
# se é <= 0,10 (10%): O índice aleatório está representando na tabela abaixo:

#dimensão da matriz/ coerencia aleatória:
#1/0,00, 2/0,00, 3/0,58, 4/0,90, 5//1,12, 6/1,24, 7/1,32, 8/1,41, 9/1,45, 10/1,49

#Neste caso o IA será 0,90, já que estamos trablhando com uma matriz 4x4 (n=4 -> numero de criterios)

#Contudo a comparação temos:
# RC = IC/IA
#tol = tolerancia para ser coerente ( 0.10)


tol = .1
IA = .9
RC = IC/IA
print(RC)

if RC <= tol:

    print('O valor', RC, ' é menor que', tol, ',ou seja verifica')

elif RC >= tol:

    print('O valor' ,RC,' é maior que', tol, ',ou seja não verifica')
```

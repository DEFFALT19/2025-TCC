## 3) Forma de aplicação do AHP

```python
# O resultado obtido pela aplicação do método AHP ocorre por meio da
# comparação paritária dos critérios, subcritérios e alternativas, quantificando os 
# fatores. Tais comparações nos permitem melhorar a consistência das decisões, com
# a utilização das informações disponíveis (SAATY, 2001). Ainda segundo o mesmo 
# autor, a tomada de decisão necessita das seguintes etapas básicas:


# - planejamento
# - geração do conjunto de alternativas
# - estabelicimento de prioridades
# - escolha da melhor política, após a definição do conjunto de alternativas
# - alocação de recursos
# - determinação dos resultados
# - previsão dos sistemas
# - avaliação do desempenho
# - garantia da estabilidade do sistema
# - otimização
# - resolção dos conflitos
```

### 4.1) Síntese do resultado das avaliações dos respondentes

```python
# Ordem de colunas: Respondente, Local 1 - 6, Inconsistência
#Ordem de linhas: Respondente, 1 - 15
# Linha L1,1 e Coluna C1,1 não inclusas na matriz abaixo A
# Matriz: 15x7
A = ([.138, .133, .122, .091, .154, .362, .14],
     [.181, .159, .141, .178, .168, .173, .16],
     [.312, .167, .125, .137, .135, .124, .10],
     [.430, .172, .115, .069, .112, .102, .17],
     [.248, .085, .150, .132, .169, .217, .04],
     [.216, .219, .137, .092, .073, .264, .16],
     [.368, .052, .173, .196, .109, .101, .19],
     [.285, .073, .131, .055, .206, .251, .09],
     [.167, .386, .120, .182, .082, .080, .20],
     [.170, .226, .088, .113, .177, .227, .04],
     [.201, .160, .142, .098, .200, .199, .06],
     [.039, .175, .059, .195, .322, .210, .18],
     [.090, .151, .375, .059, .162, .164, .14],
     [.457, .037, .210, .044, .147, .105, .28],
     [.347, .098, .159, .097, .162, .370, .09])

        

A = array(A)
print(A)
print('--------------------------------------------')

print('Formato da matri:',A.shape)

print('--------------------------------------------')
AM = np.mean(A, axis = 0)
print('Media das colunas:', AM)
```

### 4.2) Conclusões

```python
# Conforme vizualiza-se na última coluna acima (médias), a inconsistência
# global ficou dentro do limite aceito pelo autor do método, que é abaixo de 0,20.
# Porém, antes de se chegar no quadro final, houve a necessidade dos questionário
# dos respondentes 4, 6, 7, 8 e 14 serem revistos, pois algumas das avaliações destes
# apresentam uma inconsistência acima de 0,20.

# Apenas o respondente 14, após dois envios e duas revisões feitas, não alterou
# suas comparações paritárias ao ponto de se chegar à consistência aceita
# pelo método. Apesar disto, o respondente foi mantido, pois a inconsistência global manteve-se 
# dentro do limite aceitável.

# A partir da análise da tabela acima, chegou-se na visualização abaixo do ranqueamento
# dos locais candidatos, com sua classificação e seus respectivos nomes.

print( '1° / Local 1 / São Dimas/Silvestre /', AM[0])
print( '2° / Local 6 / Rodoviária/INSS /', AM[5])
print( '3° / Local 5 / Pinheirinho /', AM[4])
print( '4° / Local 2 / Portal/Bombeiros /', AM[1])
print( '5° / Local 3 / Barassa/Camanducaia /', AM[2])
print( '6° / Local 4 / Garagem Prefeitura/ Pq. Linear /', AM[3])
```

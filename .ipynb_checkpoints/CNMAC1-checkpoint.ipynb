## Programação Linear - Método Simplex Revisado

```python
# Problema:  Um pequeno produtor artesanal produz cupcakes e croissants para vender em uma
# cantina escolar. A unidade de cupcake é vendida pelo valor de 3 reais e a unidade de croissant é
# vendida pelo valor de 5 reais. O cupcake tem custo de 1 real em matéria-prima e 1 real em 
# recursos humanos para a produção. O croissant tem custo de 1 real em matéria-prima e 2 reais
# em recursos humanos. A produção destes alimentos é dividida em 2 etapas: tempo de 
# preparação da massa e tempo no forno. A produção de 10 cupcakes leva 2 horas na preparação da
# massa e 1 hora no forno. A produção de 5 croissants leva 1 hora de preparação de massa e 2 horas
# no forno. Se durante um dia de trabalho estão disponíveis 6 horas para preparação da massa e 6
# horas para o forno, utilize o método simplex revisado para determinar a quantidade de 
# cupcakes e croissant que devem ser produzidos para que se obtenha o maior lucro.
```

```python
#V1)

# Max{ z = 3x1 + 5x2 - 1x1 - 1x1 -1x2 - 2x2}
# Sujeito a
# 2/10 (horas/cupcakes) x1 + 1/5 x2 <= 6 (Tempo de preparação)
# 1/10 x1 + 2/5 (horas/croissants) x2 <= 6 (Tempo no forno)
```

```python
#V2)

# -Min{ -z = -x1 - 2x2}
# Sujeito a
# x1 + x2 <= 30 (* 5)
# x1 + 4x2 <= 60 (* 10)
```

```python
#V3)

# -Min{ -z = -x1 - 2x2}
# Sujeito a
# x1 + x2 + x3 = 30 
# x1 + 4x2 + x4 = 60 

#     x1  x2 x3 x4
C = (-1, -2, 0, 0)
C = array(C)


A = [(1, 1, 1, 0),(1, 4, 0, 1)]
A = array(A)


#A = [1, 1, 1, 0]
#    [1, 4, 0, 1]
# X = [x1, x2, x3, x4]
X = (0, 0, 30, 60)
X = array(X)

Base = (3, 4) #3 -> x3 e 4 -> x4
B = [(1,0),(0,1)]
Non = (1, 2) #1 -> x1 e 2 -> x2

Base = array(B)
Non = array(Non)
B = array(B)

# B1 = B (B1 == B^-1)

Bi = B
print(A, A[1][1])
A.shape
```

### 1° Iteração

```python
#1) C^t B (custos base: x3, x4) * B^-1

Ctb = (0,0)
Ctb = array(Ctb)

Ctb_Bi = dot(Ctb, Bi)
Ctb_Bi
```

```python
#2) C^k = Ck - (C^t B * B^-1) = Ctb_Bi * Ak (k = Non: 1 (x1) e 2 (x2))
# Menor C^k entra

Ck = []

#C^1
#Ck1 = C1 - Ctb_Bi * Ak1 (Parte x1 em A / coluna)

A1 = array([(A[0][0]), (A[0][1])])
Ck1 = C[0] - dot(Ctb_Bi,A1)
Ck1

#Ck2 = C2 - Ctb_Bi * Ak2 (Parte x2 em A / coluna)
A2 = array([(A[0][1]), (A[1][1])])
Ck2 = C[1] - dot(Ctb_Bi,A2)
Ck2

Ck.append(Ck1)
Ck.append(Ck2)
Ck = array(Ck)

print(A2)
print('O minimo é',min(Ck),', do k=2 (x2), logo x2 é candidato a entrar a base B.')
```

```python
#3) Ck  > 0?

if min(Ck) > 0:
    print('Solução ótima')

else: 
    print('Não é solução ótima, prossiga')
```

```python
#4) dB = -B^-1Ak (k = 2, xn que entra)

dB = - Bi * A2

print('d3 =', dB[0][0],'e d4 =', dB[1][1],'. d2 = (d1=0, d2=1, d3=-1, d4=-4)')

d = [0, 1, -1, -4]
```

```python
#5) lambda = l, lB(n) = lk = - xn/dn -> min(lB) = qual xB sai da base

l = []

#lB (1) = l3

l3 = -(X[2]/d[2])
l3

l4 = -(X[3]/d[3])
l4

l.append(l3)
l.append(l4)
l

min(l)

print('O minimo é',min(l),', do d4 (x4), logo x4 sai da base B.')

lb = min(l)
print(l, lb)
```

```python
#6) 
#X' (X' == Xa) = X + lb*d

Xa = X + dot(lb,d)
print(Xa)

# Z = C * Xa

Z = C * Xa
print(Z)

print('Com a nova base com a nova estrutura', Xa,' de quanto de cada xn será feito, tem um custo de Z =', min(Z))
```

```python
#7) Nova inversa
# Bi = B*Ak (Coluna de A de k que entra, neste caso x2)

#[ 1 0 | 1 ] -> [1 0 | 1 ]   ->  [1 -1/4 | 0]  =  Bi
#[0 1 | 4 ]     [0 1/4 | 1]      [0 1/4 | 1]
 



Bi = array(([1, -1/4],[0, 1/4]))
Bi
```

### 2° Iteração

## Imports & Defs

```python
import numpy as np
import sympy as sp
from sympy import symbols
import matplotlib.pyplot as plt
```

```python
def frac(x,y):
    return sp.Rational(x,y)

def array(I):

  return np.array(I)

def cross(u,v):

  return np.cross(u,v)

def dot(u,v):

    return np.dot(u,v)

def max(A):
    return np.max(A)

def mean(A):
    return np.mean(A)

def somaC(matriz):
    return np.sum(matriz, axis=0)

def somaL(matriz):
    return np.sum(matriz, axis=1)

# Definida as matrizes de comparação, o vetor de prioridade relativa é determinado por um auto
# vetor que viabilizará a determinação do grau de importância dos elementos em cada nível
# hierárquico, sendo esta a fase de priorização dos elementos do sistema.

#NC = Normalizar coluna
#MN = Matriz Normalizada

def NC(M):
    # Calcula a soma de cada coluna
    soma_colunas = somaC(M)
    
    # Divide cada elemento da coluna pela soma da coluna correspondente
    matriz_normalizada = M / soma_colunas
    
    return matriz_normalizada

#reshape (n, ) -> (n,1):

def reshape(array):
    return array.reshape(-1, 1)

def round(a,n):
    return np.round(a,n)



```

```python
# No artigo, os autores realizeram para veriifcar o peso de prefeência dos subcritérios em peso global (levando em conta os subcriterios do outro critério) e local (apenas subcriterios do mesmo criterio). NO artigo, não foi mostrado
# os pesos dados a cada subcriterio em relação aos outros nem a tabela de comparação para nos baseas, apenas o resultado final. Com isso podemos tentar replicar algo semelhante:



# OBS: Como não sabemos qual o nível de preferência de um subcritério a outro, colocarei os pesos que acho apropriado.
```

### Local (Critério 1)

```python
# Ordem de linha e colunas: 1.1 | 1.2

# MN = matriz natural, MR = matriz resultado, MM = matriz media


M1 = ([1, 1/8],
      [8, 1])

M1 = array(M1)
M1.shape

MN1 = NC(M1)
MM1 = somaL(MN1)/2
#MM1 = reshape(MM1) 
MR1 = dot(MN1, MM1)
print(MN1, MM1)
print(MM1.shape)

#row_vector = array.reshape(1, -1)
#print(row_vector.shape)

#for i in MR1:
   # print(i)

print('No critério 1, o subcritério 1.1 tem preferência de ', round(MR1[0]*100,2), '%, e o subcritério 1.2 de', round(MR1[1]*100,2), '%.')
```

```python
# teste

row_vector = MM1.reshape(-1, 1)
print(row_vector.shape)

print(dot(MN1, row_vector))
print(round(row_vector[0],2))
```

### Local (Critério 2)

```python
# Ordem de linha e colunas: 2.1 | 2.2 | 2.3

M2 = ([1, 8, 4],
      [1/8, 1, 5],
      [1/4, 1/5, 1])

M2 = array(M2)

MN2 = NC(M2)
MM2 = somaL(MN2)/3
MR2 = dot(MN2, MM2)
print(MR2)

print('No critério 2, o subcritério 2.1 tem preferência de ', round(MR2[0]*100,2), '%, o subcritério 2.2 de', round(MR2[1]*100,2), '%, e o subcritério 2.3 de', round(MR2[2]*100,2), '%.')
```

```python
#Teste

print(MN2.shape)
print(MM2.shape)
a = reshape(MM2)
a.shape
```

### Global

```python
# Ordem de linha e colunas: 1.1 | 1.2 | 2.1 | 2.2 | 2.3

# Matriz de preferencia = MP

MP = ([MR1[0]], [MR1[1]], [MR2[0]], [MR2[1]], [MR2[2]])
MP = array(MP)
print(MP.shape)


# Matriz = M

M = ([1, 1/8, 1/6, 4, 4],
     [8, 1, 1/3, 7, 8],
     [6, 3, 1, 8, 4],
     [1/4, 1/7, 1/8, 1, 5],
     [1/4, 1/8, 1/4, 1/5, 1])

M = array(M)

print(M.shape)
print(M)
```

```python
MN = NC(M)
print(MN)
print(MN.shape)
```

```python
# Matriz média (MM) dos critérios:

MM = somaL(MN)/5
print(MM.shape)
print(MM[2])
print(MM)
print(sum(MM))
```

```python
#Matriz resposta

MR = dot(MN,MM)
print(MR)


c1 = MR[0] + MR[1]
c2 = MR[2] + MR[3] + MR[4]

print(c1, c2)

print(round(MR[4],4)*100)
```

```python
# Resultado

print('No caso de comparação global, concluimos que o subcritérios 1.1 e 1.2 tem preferência de', round(MR[0],2)*100, '% e ', round(MR[1],2)*100, '% respectivamente.'
      ' Por outro lado, os subcritérios 2.1, 2.2 e 2.3 tem preferência de', round(MR[2],2)*100, '%,', round(MR[3],2)*100,'% e', round(MR[4],4)*100, '% respectivamente.'
      ' Em relação aos critérios, o critério 1 tem preferência de', round(c1,2)*100, '%, e o critério 2 com preferência de', round(c2,2)*100,'%.'
     
     
     
     
     )
```

# Imports & Defs

```python
import numpy as np
import sympy as sp
from sympy import symbols
```

```python
def array(I):

  return np.array(I)

def dot(u,v):

    return np.dot(u,v)

def max(A):
    return np.max(A)

def mean(A):
    return np.mean(A)

def somaC(matriz):
    return np.sum(matriz, axis=0)

def somaL(matriz):
    return np.sum(matriz, axis=1)

# Definida as matrizes de comparação, o vetor de prioridade relativa é determinado por um auto
# vetor que viabilizará a determinação do grau de importância dos elementos em cada nível
# hierárquico, sendo esta a fase de priorização dos elementos do sistema.

#NC = Normalizar coluna
#MN = Matriz Normalizada

def NC(M):
    # Calcula a soma de cada coluna
    soma_colunas = somaC(M)
    
    # Divide cada elemento da coluna pela soma da coluna correspondente
    matriz_normalizada = M / soma_colunas
    
    return matriz_normalizada

def round(a,n):
    return np.round(a,n)


def append(X, a):

    return np.append(X,a)
```

### Criterio 1

```python
#Criterio 1: Frequencia
#1.1- Tempo de resposta
#1.2- N de ocorrência
#1.3- grau de importancia

C1 = ([1  , 9,   2],
      [1/9, 1, 1/9],
      [1/2, 9,   1])

C1=array(C1)

C1.shape
C1

```

```python
# Normalização

C1N = NC(C1)
C1N
```

```python
#Vetor media/prioridade

VM1 = somaL(C1N)/3
print(sum(VM1))
print(VM1)
VM1.shape

VM1*100
```

```python
# Teste de consistencia

lmax = sum(dot(C1,VM1)/VM1)/3
CI = (lmax-3)/(3-1)
IA = .58
CR = CI/IA
print(CR)
```

```python
#Calcular o indice de coerencia IC (CI): (lmax - n)/(n - 1)


#1) Determinar o lambda máximo, considera-se o n° de subcritérios. Divide-se o vetor do preferência de entradas (Mx) pelo vector da Média (MMx) e obtêm-se um novo vetor -> lmax = sum(dot(Mx,MM)/MM)/n.
# Deste vetor resultante, somam-se as suas parcelas e divide-se pelo n° de subcritérios, que resulta no lambda máximo:

CI = (lmax-3)/(3-1)
CI
```

```python
# Comparar indice de coerencia IC ao indice aleatorio IA, e verificar se RC <= 0.1 (10%)

IA = .58
RC = CI/IA
print(RC)
```
